# Welcome to an example Binder

We need `seaborn`, which shall be included it in `requirements.txt` file

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
df=pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
model=sm.OLS.from_formula('diastol ~ sistol',data= df)
result=model.fit()
result.summary()
```

```python
model=sm.OLS.from_formula('peso ~ cadera',data= df)
result=model.fit()
result.summary()
```

```python
model=sm.OLS.from_formula('peso ~ cadera',data= df)
result=model.fit()
result.summary()
```

```python
model=sm.OLS.from_formula('diastol ~ sistol + edad + peso + cintura',data= df)
result=model.fit()
result.summary()
```

```python
import statistics

statistics.median(result.resid)
```
