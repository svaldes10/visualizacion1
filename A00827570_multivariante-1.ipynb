```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import pandas as pd


datos = "datasets/antropometria-dataset.csv"

df = pd.read_csv(datos)

df.head()
```

```python
df.fillna(method ='ffill', inplace = True) 

peso = df['peso']
cintura = df['cintura']
```

```python
plt.figure(figsize=(10,5))


plt.subplot(1,2,1)
plt.hist(x = peso, bins = 5)
plt.title("Peso")

plt.subplot(1,2,2)
plt.hist(x = cintura, bins = 5)
plt.title("Cintura")

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x = peso, y = cintura)
plt.xlabel("'peso'")
plt.ylabel("'cintura'")
plt.title("Diagrama de dispersiÃ³n de peso vs. cintura")

plt.show()
```
