```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "datasets/AsiaCases_.csv"
df = pd.read_csv(url)
type(df)
```

```python
print(df.columns)

df.drop(['ID', 'Country','TotalCasesPerMillion','TotalDeathsPerMillion','TotalTestsPerMillion'], axis=1, inplace=True)
print(df.columns)
df.head()
```

```python
df.set_index('TotalCases')
```

```python
df.describe()
```

```python
sns.scatterplot(x = 'TotalDeaths', y = 'TotalCases', data = df)
sns.set(rc={'figure.figsize':(30,11)})
plt.title('Total de Casos Covid comparados con Total de Muertes')
plt.show()
```
