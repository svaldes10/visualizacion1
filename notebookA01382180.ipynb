```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn import linear_model
```

```python
#¿Qué tanto afectan las variables en las victorias de los equipos?
#correlacion y regresión 
```

```python
regresion = sm.OLS.from_formula('W ~ ADJOE + ADJDE + TOR + TORD + ORB + FTR', data = df)
resultado = regresion.fit()
resultado.summary()
```

```python
#¿Como afecta cada variable por separado a las vistorias de los equipos?
```

```python
#¿Entre mejor Rank de la ofensiva mas victorias del equipo?
from statsmodels.graphics.regressionplots import plot_ccpr

plot1 = plt.axes()
plot_ccpr(resultado, "ADJOE", plot1)
_ = plot1.lines[0].set_alpha(0.2)
_ = plot1.set_ylabel("W")
_ = plot1.set_title('ADJOE values')
```

```python
#Se puede observar que la variable del ADJOE de la ofensiva es proporcional a mas victorias y tiene una relacion lienal positiva lo que significa que si tu rank es de los primeros tendras menos victorias y viceversa.
```
