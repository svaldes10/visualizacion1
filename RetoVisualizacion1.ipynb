```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "datasets/top250-00-19.csv"
df = pd.read_csv(url)
type(df)
```

```python
print(df.columns)

df.drop(['Position', 'Name'], axis=1, inplace=True)
print(df.columns)
df.head()
```

```python
df.set_index('Age')
```

```python
df.describe()
```

```python
x = df['Age'].value_counts()
x
```

```python
x.plot(kind = "bar")
```

```python
x.plot(kind = "barh")
```

```python
sns.scatterplot(x = 'Age', y = 'League_to', data = df)
sns.set(rc={'figure.figsize':(20,11)})
plt.title('Edad en base a la futura liga')
plt.show()
```

```python
df['Age'].plot(kind = 'hist',
              figsize =(20,10))
```

```python
df['Transfer_fee'] = df['Transfer_fee'].astype(int)
df.corr()
```

```python
sns.lmplot(x='Age',
          y='Market_value',
          data=df,
          fit_reg=True,
          )
plt.title('Age vs Market Value')
```

```python
sns.boxplot(data=df.loc[:,['Age']])
```
